```python
#Write a program that takes marks of N subjects as input from the user.
subjects=int(input("Enter number of subjects:"))
total=0
for i in range(subjects):
    inp=int(input(f"subject marks {i+1} :"))
    if inp < 0 or inp >100:
        print("invalid marks!give marks between 0 to 100" )
        continue
    total=total+inp
average = total/subjects
print("enter total marks:",total)
print("average marks:",average)                  
```
